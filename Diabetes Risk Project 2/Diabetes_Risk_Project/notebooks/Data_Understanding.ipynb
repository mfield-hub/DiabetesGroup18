```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/Diabetes_and_LifeStyle_Dataset_.csv")
df.head()
```

```python
df['diabetes_stage'].value_counts()
```

```python
plt.figure()
df['diabetes_stage'].value_counts().plot(kind='bar')
plt.title("Distribution of Diabetes Stage")
plt.xlabel("Diabetes Stage")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df.hist(figsize=(14,10))
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(10,8))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr())

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='diabetes_stage', y='bmi', data=df)
plt.title("BMI vs Diabetes Stage")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='diabetes_stage', y='Age', data=df)
plt.title("Age vs Diabetes Stage")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='diabetes_stage', y='physical_activity_minutes_per_week', data=df)
plt.title("Physical Activity vs Diabetes Stage")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Outlier Detection")
plt.show()
```
